```python
import numpy as np
from scipy import integrate
```

 ЗАВДАННЯ 1: Метод прямокутників

 Інтеграл: 1 / sqrt(x + 2.5) від 1.6 до 2.2

 n = 10Завдання 1. Метод прямокутників (правих, лівих і середніх)

```python
def f1(x):
    return 1 / np.sqrt(x + 2.5)

a1 = 1.6
b1 = 2.2
n1 = 10
h1 = (b1 - a1) / n1

sum_left = sum([f1(a1 + i * h1) for i in range(n1)])
result_left = h1 * sum_left

sum_right = sum([f1(a1 + (i + 1) * h1) for i in range(n1)])
result_right = h1 * sum_right

sum_mid = sum([f1(a1 + (i + 0.5) * h1) for i in range(n1)])
result_mid = h1 * sum_mid

print(f"--- ЗАВДАННЯ 1 (n={n1}) ---")
print(f"Метод лівих прямокутників:   {result_left:.5f}")
print(f"Метод правих прямокутників:  {result_right:.5f}")
print(f"Метод середніх прямокутників:{result_mid:.5f}")

check1, err1 = integrate.quad(f1, a1, b1)
print(f"Перевірка (SciPy):           {check1:.5f}")
print("-" * 30)
```

 ЗАВДАННЯ 2: Метод Сімпсона

 Інтеграл: (x^2 + 1) * sin(x - 0.5) від 1.3 до 2.1

 n = 8


```python
def f2(x):
    return (x**2 + 1) * np.sin(x - 0.5)

a2 = 1.3
b2 = 2.1
n2 = 8
h2 = (b2 - a2) / n2

x_vals = [a2 + i * h2 for i in range(n2 + 1)]
y_vals = [f2(x) for x in x_vals]

simpson_sum = y_vals[0] + y_vals[-1]
simpson_sum += 4 * sum(y_vals[1:-1:2])
simpson_sum += 2 * sum(y_vals[2:-1:2])

result_simpson = (h2 / 3) * simpson_sum

print(f"--- ЗАВДАННЯ 2 (n={n2}) ---")
print(f"Метод Сімпсона:    {result_simpson:.5f}")

check2, err2 = integrate.quad(f2, a2, b2)
print(f"Перевірка (SciPy): {check2:.5f}")
print("-" * 30)
```

 ЗАВДАННЯ 3: Метод трапецій

 Інтеграл: 1 / sqrt(x^2 + 4) від 0.8 до 1.8

 n = 20

```python
def f3(x):
    return 1 / np.sqrt(x**2 + 4)

a3 = 0.8
b3 = 1.8
n3 = 20
h3 = (b3 - a3) / n3

y0 = f3(a3)
yn = f3(b3)
sum_rest = sum([f3(a3 + i * h3) for i in range(1, n3)])

result_trapezoid = h3 * ((y0 + yn) / 2 + sum_rest)

print(f"--- ЗАВДАННЯ 3 (n={n3}) ---")
print(f"Метод трапецій:    {result_trapezoid:.5f}")

check3, err3 = integrate.quad(f3, a3, b3)
print(f"Перевірка (SciPy): {check3:.5f}")
print("-" * 30)
```
